```python
import numpy as np
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%".format(pct, absolute)
groupe=data["Groupe"].value_counts()
g=data.groupby(['Groupe']).size()
g.plot.pie(autopct = lambda pct: func(pct, groupe))
```
