# visualisation

```python
karate = nx.karate_club_graph()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nx.draw(karate)
```

```python
nx.draw(karate, with_labels=True)
```

```python
plt.figure(figsize=(10,10))
#multu dimension 
plt.subplot(221)  
nx.draw_spring(karate, with_labels=True)
plt.subplot(222) 
nx.draw_spectral(karate, with_labels=True)
plt.subplot(223) 
nx.draw_circular(karate, with_labels=True)
plt.subplot(224) 
nx.draw_kamada_kawai(karate, with_labels=True)
```

# statisc de graph

```python
print('la densite :', nx.density(karate))
```

```python
print('la plus courte longeur ', nx.average_shortest_path_length(karate))
```

```python
print('clustering coefficient:', nx.average_clustering(karate))
```
