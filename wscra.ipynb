```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, 'html.parser')
```

```python
title=soup.find('title')
```

```python
items=soup.findAll('item')
```

```python
listes=soup.find_all("section", class_="listing-search-item")
```

```python
for liste in listes : 
    title=liste.find('a',class_="listing-search-item__link--title").text.replace('\n', '') 
    location= liste.find('div', class_="listing-search-item__sub-title").text.replace('\n', '') 
    price= liste.find('div', class_="listing-search-item__price").text.replace('\n', '') 
    area= liste.find('li', class_="illustrated-features__item illustrated-features__item--surface-area").text.replace('\n', '') 
    information =[title, location, price, area] 
    print(information)
```

```python
from csv import writer
with open('appart.csv', 'w', encoding='UTF8', newline='') as f:
    thewriter= writer(f)
    header= ['Title', 'Location', 'Price', 'Area']
    thewriter.writerow(header)
    for liste in listes :
        title=liste.find('a',class_="listing-search-item__link--title").text.replace('\n', '')
        location= liste.find('div', class_="listing-search-item__sub-title").text.replace('\n', '')
        price= liste.find('div', class_="listing-search-item__price").text.replace('\n', '')
        area= liste.find('li', class_="illustrated-features__item illustrated-features__item--surface-area").text.replace('\n', '')
        information =[title, location, price, area]
        thewriter.writerow(information)
```
