# Visualisation univari√©e :

```python
sns.distplot(data["Model"])
```

```python
sns.kdeplot(data["Model"])
```

```python
sns.histplot(data.Model)
```

# Visualisation des boites √† moustaches

```python
sns.boxplot(x =data.Model, y =data.MPG, data = data)
```

```python
sns.violinplot(x =data.Model, y =data.MPG, data = data)
```

# #Nuage de points :

```python
sns.scatterplot(data=data, x="Model", y="MPG")
```

```python
sns.scatterplot(data=data, x="Horsepower", y="Displacement",hue="Model",palette="deep")
```

```python
sns.relplot( data=data, x="Horsepower", y="Displacement",kind="scatter" )
```

```python
sns.stripplot(data=data, y="Horsepower",x='Displacement', hue='Model')
```

```python
sns.swarmplot(data=data, x="Cylinders",y='Acceleration', hue='Model',legend=False)
```

```python
sns.catplot( data=data, kind="swarm", x="Horsepower", y="Acceleration" )
```
