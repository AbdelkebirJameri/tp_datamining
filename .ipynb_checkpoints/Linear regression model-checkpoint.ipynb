```python
x=df.drop("Y", axis=1)
y=df['Y']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 101) 
```

```python
from sklearn import  linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(x_train, y_train)  


y_pre = model.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pre)
```

```python
mse=mean_squared_error(y_test, y_pre)
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()

# Fit the model
model_lasso1 = lasso.fit(x_train, y_train)

# Prediction
y_pred_train_lass = lasso.predict(x_train)
y_pred_test_lass = lasso.predict(x_test)

# Accuracy Score
print('Training accuracy : {}\n'.format(r2_score(y_train, y_pred_train_lass).round(5)))
print('Testing accuracy : {}'.format(r2_score(y_test, y_pred_test_lass).round(5)))
```

```python
 from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()

# Fit the model
model_ridge = ridge.fit(x_train, y_train)

# Prediction
y_pred_train_ridge = ridge.predict(x_train)
y_pred_test_ridge = ridge.predict(x_test)

# Accuracy Score
print('Training accuracy : {}\n'.format(r2_score(y_train, y_pred_train_ridge).round(5)))
print('Testing accuracy : {}'.format(r2_score(y_test, y_pred_test_ridge).round(5)))
```
